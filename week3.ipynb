## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier as MLP
import warnings
warnings.filterwarnings('ignore')
```

# 1. Multi Layer Perceptron Data XOR

```python
dataset = pd.read_csv('../Data/XOR.csv')
data = dataset.iloc[:,0:-1]
label = dataset.iloc[:,-1]
iter = np.array([10, 20, 30, 40, 50, 100, 150 , 200, 250, 300])
def mse(actual, pred):
    actual, pred = np.array(actual), np.array(pred)
    return np.square(np.subtract(actual,pred)).mean()

for i in range (0,10):
    model = MLP(hidden_layer_sizes=(7), max_iter=iter[i], activation = 'logistic',
                learning_rate_init=0.1)
    model.fit(data,label)
    pred=model.predict(data)
    if i==0:
        error = np.array(mse(label, pred))
    else:
        error = np.append(error, mse(label, pred))
    print(f'accuracy score with {iter[i]} iterations\t: {model.score(data,label)}')

plt.plot(iter,error)
plt.xlabel('Iteration No.')
plt.ylabel('Mean Square Error')
plt.title('Gradient Descent XOR')
plt.show()
```

# 4. Melakukan testing untuk data testing
## a. Data Ruspini

```python
dataset = pd.read_csv("../Data/ruspini.csv")
data = dataset.iloc[:, 1:-1]
label = dataset.iloc[:, -1]
iter = np.array([10, 20, 30, 40, 50, 100, 150, 200, 250, 300, 450, 500])

def mse(actual, pred):
    actual, pred = np.array(actual), np.array(pred)
    return np.square(np.subtract(actual, pred)).mean()

data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=0.2, random_state=42)

error_train = []
error_test = []

for i in range(12):
    model = MLP(hidden_layer_sizes=(301), max_iter=iter[i], activation = 'logistic', solver='lbfgs',
                learning_rate_init=0.1, random_state=42)
    model.fit(data_train, label_train)
    
    pred_train = model.predict(data_train)
    pred_test = model.predict(data_test)
    
    error_train.append(mse(label_train, pred_train))
    error_test.append(mse(label_test, pred_test))

    print(f'iteration: {iter[i]}, train Score: {model.score(data_train, label_train):.2f}, test Score: {model.score(data_test, label_test):.2f}')

plt.plot(iter, error_train, label='Training MSE')
plt.plot(iter, error_test, label='Testing MSE')
plt.xlabel('Iteration No.')
plt.ylabel('Mean Square Error')
plt.title('MSE for Training and Testing Data (Dataset Ruspini)')
plt.legend()
plt.show()
```
