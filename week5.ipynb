## import required libraries

````python
import math
import matplotlib.pyplot as plt
from importlib import reload
import pandas as pd
from sklearn.neural_network import MLPClassifier as MLP
import warnings
warnings.filterwarnings('ignore') 
````

## read dataset digit.csv

````python
dataset = pd.read_csv('../Data/digit.csv')
data = dataset.iloc[:, 0:-1]
label = dataset.iloc[:, -1]
dataset.head()
````

## data visualization

````python
numbers_to_display = 80
num_cells = math.ceil(math.sqrt(numbers_to_display))
plt.figure(figsize=(20, 20))
for i in range(numbers_to_display):
    ax = plt.subplot(num_cells, num_cells, i + 1)
    im = ax.imshow(data.iloc[i].values.reshape(10, 10))
    plt.xlabel = ''
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    plt.title(label.iloc[i])
````

## MLP model

````python
model = MLP(tol = 0, hidden_layer_sizes = (10), activation = 'logistic', learning_rate_init = 0.1)
model.fit(data, label)
````

## score evaluation

````python
print('score:', model.score(data, label))
print('predictions:', model.predict(data))
print('expected:', label)
````

## plot gradient descent

````python
plt = reload(plt)
plt.plot(model.loss_curve_)
plt.xlabel('Iterations')
plt.ylabel('loss')
plt.title('Gradient Descent')
plt.show()
````

### 2. Buatlah klasifikasi Neural Network dengan menjadikan dataset digit yang baru menjadi data testing!

````python
dataset = pd.read_csv('../Data/digit.csv')
test_data = dataset.iloc[:, 0:-1]
model = MLP(tol = 0, hidden_layer_sizes = (10), activation = 'logistic', learning_rate_init = 0.1)
model.fit(test_data, label)

print('score:', model.score(test_data, label))
print('predictions:', model.predict(test_data))
````

````python
plt = reload(plt)
plt.plot(model.loss_curve_)
plt.xlabel('Iterations')
plt.ylabel('loss')
plt.title('Gradient Descent')
plt.show()
````

### 3. Lakukan dengan menggunakan ```miu = 0.1, dan layer = 3, 5, 10, 15, 20, dan 25```

````python
hidden_layers = [3, 5, 10, 15, 20, 25]

plt.figure(figsize=(10, 6))
for layer_size in hidden_layers:
    model = MLP(hidden_layer_sizes=(layer_size,), activation='logistic',
                learning_rate_init=0.1, random_state=42)
    model.fit(data, label)
    pred = model.predict(data)
    plt.plot(model.loss_curve_, label=f'Layer Size: {layer_size}')

plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Gradient Descent Loss Curves for Various Hidden Layer Sizes with Learning Rate 0.1')
plt.legend()
plt.show()
````

### 4. Lakukan dengan menggunakan ```layer = 10 dengan miu = 0.1, 0.5, 3, 5, 8, dan 10```

````python
learning_rates = [0.1, 0.5, 3, 5, 8, 10]

plt.figure(figsize=(10, 6))
for lr in learning_rates:
    model = MLP(hidden_layer_sizes=(10,), activation='logistic',
                learning_rate_init=lr, random_state=42)
    model.fit(data, label)
    
    pred = model.predict(data)
    
    plt.plot(model.loss_curve_, label=f'Learning Rate: {lr}')

plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Gradient Descent Loss Curves for Various Learning Rates with Layer Size 10')
plt.legend()
plt.show()
````
